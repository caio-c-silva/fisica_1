$\ddot{x} = -\frac{c}{m}\sqrt{{\dot{x}}^2 + {\dot{y}}^2}\dot{x}$

$\ddot{y} = -\frac{c}{m}\sqrt{{\dot{x}}^2 + {\dot{y}}^2}\dot{y} - g$

Para resolver essa equação diferencial, utilizamos as seguintes definições:

$u_1 \equiv x$, $u_2 \equiv \dot{x}$, $u_3 \equiv y$ e $u_4 \equiv \dot{y}$. Assim, ao ivés de 2 equações de segunda ordem, temos 4 equações de primeira ordem acopladas:

$\dot{u_1} = u_2$

$\dot{u_2} = -\frac{c}{m}\sqrt{{\dot{x}}^2 + {\dot{y}}^2}\dot{x}$

$\dot{u_3} = u_4$

$\dot{u_4} = -\frac{c}{m}\sqrt{{\dot{x}}^2 + {\dot{y}}^2}\dot{y} - g$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Coeficiente de Resitencia do Ar, Raio do Projétil (m), Área (m2) and Massa (kg).
c = 0.47
r = 0.05
A = np.pi * r**2
m = 0.2
# Densidade do Ar (kg.m-3), Aceleração da Gravidade (m.s-2).
rho_ar = 1.28
g = 9.81

# Definir constante de resistência.
#k = 0.5 * c * rho_ar * A

# Sem Resistência do Ar 
k=0

# Velocidade Inicial e Âgulo de Lançamento.
v0 = 10

theta_d1 = 15
theta_d2 = 45
theta_d3 = 75


theta1 = np.radians(theta_d1)
theta2 = np.radians(theta_d2)
theta3 = np.radians(theta_d3)
```

```python
def deriv(t, u):
    x, xdot, y, ydot = u
    speed = np.hypot(xdot, ydot)
    xdotdot = -k/m * speed * xdot
    ydotdot = -k/m * speed * ydot - g
    return xdot, xdotdot, ydot, ydotdot
```

```python
# Condições Iniciais: x0, v0_x, y0, v0_y.
u01 = 0, v0 * np.cos(theta1), 0., v0 * np.sin(theta1)
u02 = 0, v0 * np.cos(theta2), 0., v0 * np.sin(theta2)
u03 = 0, v0 * np.cos(theta3), 0., v0 * np.sin(theta3)

# Integrar até tf, a menos que o alvo seja alcançado antes.
t0, tf = 0, 100
```

```python
def atingir_alvo(t, u):
    # Alvo atingido quando y = 0.
    return u[2]

# Parar a integração quando o alvo for atingido.
atingir_alvo.terminal = True

# Parar na descida (não parar na subida)
atingir_alvo.direction = -1

def altura_maxima(t, u):
    # A altura máxima acontece quando a velocidade em y é igual a  0.
    return u[3]
```

```python
soln1 = solve_ivp(deriv, (t0, tf), u01, dense_output=True,
                 events=(atingir_alvo, altura_maxima))
soln2 = solve_ivp(deriv, (t0, tf), u02, dense_output=True,
                 events=(atingir_alvo, altura_maxima))
soln3 = solve_ivp(deriv, (t0, tf), u03, dense_output=True,
                 events=(atingir_alvo, altura_maxima))


# Mais pontos em t.
t1 = np.linspace(0, soln1.t_events[0][0], 100)
t2 = np.linspace(0, soln2.t_events[0][0], 100)
t3 = np.linspace(0, soln3.t_events[0][0], 100)

# Solução para o gráfico da tragetória.
sol1 = soln1.sol(t1)
x1, y1 = sol1[0], sol1[2]

sol2 = soln2.sol(t2)
x2, y2 = sol2[0], sol2[2]

sol3 = soln3.sol(t3)
x3, y3 = sol3[0], sol3[2]
```

```python
plt.plot(x1, y1, label=fr'$\theta_0={theta_d1}^{{\circ}}$')
plt.plot(x2, y2, label=fr'$\theta_0={theta_d2}^{{\circ}}$')
plt.plot(x3, y3, label=fr'$\theta_0={theta_d3}^{{\circ}}$')
plt.ylim(bottom=0)
plt.legend()
plt.xlabel('$x (m)$', fontsize=20)
plt.ylabel('$y (m)$', fontsize=20)
plt.show()
```

```python
print(f'Para Theta = {theta_d1}')
print('')

print(f'Tempo até o alvo = {round(soln1.t_events[0][0], 2)} s')
print(f'Tempo até a altura máxima = {round(soln1.t_events[1][0], 2)} s')
print(f'Distância Máxima, xmax = {round(x1[-1], 2)} m')
print(f'Altura Máxima, ymax = {round(max(y1), 2)} m')
print('')

print(f'Para Theta = {theta_d2}')
print('')

print(f'Tempo até o alvo = {round(soln2.t_events[0][0], 2)} s')
print(f'Tempo até a altura máxima = {round(soln2.t_events[1][0], 2)} s')
print(f'Distância Máxima, xmax = {round(x2[-1], 2)} m')
print(f'Altura Máxima, ymax = {round(max(y2), 2)} m')
print('')

print(f'Para Theta = {theta_d3}')
print('')

print(f'Tempo até o alvo = {round(soln3.t_events[0][0], 2)} s')
print(f'Tempo até a altura máxima = {round(soln3.t_events[1][0], 2)} s')
print(f'Distância Máxima, xmax = {round(x3[-1], 2)} m')
print(f'Altura Máxima, ymax = {round(max(y3), 2)} m')
```
